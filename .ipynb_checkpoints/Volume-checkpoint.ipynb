import libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
from fbprophet import Prophet
from datetime import datetime
%matplotlib inline 
```

Read the csv files into dataframes. CSV files include deposits,loans taken and all successful purchases of token between January and March 2019.
The features include:
1.address: the mobile number of account
2.amount/transamount: the total paid or borrowed for loan purchase
3.unit: the unit of the token
4.transactionid : unique identifier for each transaction
5.time/systranstime: the timestamp of the transaction
6.billrefnumber:unique identifier for account

successfulVends_raw holds data for successful purchases
deposit_raw holds data for purchases
loan_raw holds data for borrowed amounts to purchase tokens

They have been named raw because they contain unrequired features for building the model

```python
successfulVends_raw = pd.read_csv("successfulVENDS2019.csv")
deposit_raw = pd.read_csv("deposit2019.csv")
loan_raw = pd.read_csv("loan2019.csv")
```

Selecting required features

```python

deposit = deposit_raw[["billrefnumber","systemtranstime","transamount"]]
successfulVends = successfulVends_raw[["address","time","amount"]]
```

Working on deposit and SuccessfulVends first

Renaming the deposit features to amount and time

```python

deposit= deposit.rename(index=str,columns={"transamount":"amount","systemtranstime":"time"})
```

Converting time column to datetime type for easier manipulation

```python
deposit['time']=pd.to_datetime(deposit.time)
successfulVends['time']=pd.to_datetime(successfulVends.time)
```

```python
def month(df):
    for i in range(1,4):
        df[df.index.month==i].plot()
```

Raw values no resampling

```python
plt.plot(deposit.time,deposit.amount)
```

Resampled by minute 

```python
deposit_min = deposit.resample('1min', on='time').mean()
month(deposit_min)
```

Resampled by hour

```python
deposit_hour = deposit.resample('60min', on='time').sum()
month(deposit_hour)
```

Resampled by day

```python
deposit_day = deposit.resample('1d', on='time').sum()
month(deposit_day)
```

Resampled by 3H

```python
deposit_3H = deposit.resample('3H', on='time').sum()
month(deposit_3H)
```

Set the time column as index since it is a unique identifier

```python
fb_deposit = deposit.rename(index=str,columns={"amount":"y","time":"ds"})
m = Prophet()
m.fit(fb_deposit)
```

```python
future = m.make_future_dataframe(periods=5, freq='H')
```

```python
forecast = m.predict(future)
```

```python
fig1 = m.plot(forecast)
plt.ylim(0, 10000)
```

```python
fig2 = m.plot_components(forecast)
```
